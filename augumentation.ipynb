```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for training and validation Dice coefficients
epochs = np.arange(0, 400)
final_dice_value = 0.976168899015199

# Generate training Dice coefficients
train_dice = np.cumsum(np.random.normal(0.002, 0.005, size=epochs.shape))
train_dice = np.clip(train_dice, 0.5, final_dice_value)
train_dice[-1] = final_dice_value  # Ensure the final value is exactly 0.976168899015199

# Generate validation Dice coefficients
val_dice = np.cumsum(np.random.normal(0.002, 0.01, size=epochs.shape))
val_dice = np.clip(val_dice, 0.5, final_dice_value)
val_dice[-1] = final_dice_value  # Ensure the final value is exactly 0.976168899015199

# Adding some noise to mimic the fluctuations
train_dice[10:30] += np.random.normal(0, 0.05, size=20)
val_dice[10:30] += np.random.normal(0, 0.05, size=20)
val_dice[200:220] -= np.random.normal(0, 0.1, size=20)

plt.figure(figsize=(12, 6))
plt.plot(epochs, train_dice, label='train', color='blue')
plt.plot(epochs, val_dice, label='val', color='orange')

plt.xlabel('Epochs')
plt.ylabel('Dice Coefficient')
plt.title('Dice Coefficient over Epochs')
plt.legend()
plt.grid(True)
plt.ylim(0.4, 1.05)
plt.xlim(0, 400)

# Save the plot as an image
plt.savefig('dice_coefficient_plot.png')
plt.show()
```

```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

# Preprocessing function for images
def preprocess_image(image, target_size=(512, 512)):
    # Resize image
    resized_image = cv2.resize(image, target_size)

    # Apply CLAHE for contrast enhancement to each channel
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_image = np.zeros_like(resized_image)
    for i in range(3):
        enhanced_image[:, :, i] = clahe.apply(resized_image[:, :, i])

    # Normalize pixel values to [0, 1]
    normalized_image = enhanced_image / 255.0

    return normalized_image

# Preprocessing function for masks
def preprocess_mask(mask, target_size=(512, 512)):
    if mask is None or mask.size == 0:
        raise ValueError("Mask is empty or not read correctly.")
    # Resize mask
    resized_mask = cv2.resize(mask, target_size, interpolation=cv2.INTER_NEAREST)
    # Normalize mask values to [0, 1]
    normalized_mask = resized_mask / 255.0
    return normalized_mask

# Define augmentation transforms
augmentation_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(degrees=45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    transforms.ToTensor()
])

# Custom dataset class
class SegmentationDataset(Dataset):
    def __init__(self, image_folder, mask_folders, target_size=(512, 512), augment=False):
        self.image_folder = image_folder
        self.mask_folders = mask_folders
        self.target_size = target_size
        self.augment = augment
        self.image_names = [f for f in sorted(os.listdir(image_folder)) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff'))]
        self.num_lesions = len(mask_folders)

    def __len__(self):
        return len(self.image_names)

    def __getitem__(self, idx):
        img_name = self.image_names[idx]
        img_path = os.path.join(self.image_folder, img_name)
        img = cv2.imread(img_path)
        if img is None:
            raise ValueError(f"Image {img_path} not read correctly.")
        preprocessed_img = preprocess_image(img, self.target_size)

        masks = np.zeros((self.target_size[0], self.target_size[1], self.num_lesions))
        for lesion_index, lesion_folder in enumerate(self.mask_folders):
            mask_name = img_name.replace('.png', '.png')  # Ensure the mask is in PNG format
            mask_path = os.path.join(lesion_folder, mask_name)
            if not os.path.exists(mask_path):
                raise FileNotFoundError(f"Mask file {mask_path} does not exist.")
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)  # Read as grayscale
            preprocessed_mask = preprocess_mask(mask, self.target_size)
            masks[:, :, lesion_index] = preprocessed_mask

        preprocessed_img = np.transpose(preprocessed_img, (2, 0, 1))
        masks = np.sum(masks, axis=-1, keepdims=True)
        masks = np.transpose(masks, (2, 0, 1))

        if self.augment:
            preprocessed_img = Image.fromarray((preprocessed_img.transpose(1, 2, 0) * 255).astype(np.uint8))
            masks = Image.fromarray((masks.transpose(1, 2, 0).squeeze() * 255).astype(np.uint8))
            preprocessed_img = augmentation_transforms(preprocessed_img)
            masks = augmentation_transforms(masks)

        return torch.tensor(preprocessed_img, dtype=torch.float32), torch.tensor(masks, dtype=torch.float32)

# Define paths
image_folder = "/content/drive/MyDrive/Project DR/segset/Original_Images _SoftExudate_Mask"
mask_folders = [
    "/content/drive/MyDrive/Project DR/segset/SoftExudate_Masks"
]

# Create dataset and data loaders
dataset = SegmentationDataset(image_folder, mask_folders, target_size=(512, 512), augment=True)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=2)

# Print the number of images
print(f"Number of images in the dataset: {len(dataset)}")

# Define the save path for the model checkpoints
save_path = "/content/drive/MyDrive/Project DR"

# Ensure the save path directory exists
os.makedirs(save_path, exist_ok=True)

# Visualize the augmented images and save them
def visualize_and_save_augmented_samples(dataset, idx, save_folder):
    img, mask = dataset[idx]
    img = img.numpy().transpose(1, 2, 0)  # Convert to (H, W, C)
    mask = mask.numpy().squeeze()  # Convert to (H, W)

    # Convert to PIL Image for augmentation
    img_pil = Image.fromarray((img * 255).astype(np.uint8))
    mask_pil = Image.fromarray((mask * 255).astype(np.uint8))

    # Apply each augmentation step individually
    augmented_images = [
        ("Original", img_pil, mask_pil),
        ("Horizontal Flip", transforms.RandomHorizontalFlip(p=1.0)(img_pil), transforms.RandomHorizontalFlip(p=1.0)(mask_pil)),
        ("Vertical Flip", transforms.RandomVerticalFlip(p=1.0)(img_pil), transforms.RandomVerticalFlip(p=1.0)(mask_pil)),
        ("Rotation", transforms.RandomRotation(degrees=45)(img_pil), transforms.RandomRotation(degrees=45)(mask_pil)),
        ("Color Jitter", transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2)(img_pil), mask_pil)
    ]

    # Plot and save the images
    plt.figure(figsize=(18, 12))
    for i, (title, aug_img, aug_mask) in enumerate(augmented_images):
        plt.subplot(2, len(augmented_images), i + 1)
        plt.title(f'{title} Image')
        plt.imshow(aug_img)
        plt.axis('off')

        plt.subplot(2, len(augmented_images), i + 1 + len(augmented_images))
        plt.title(f'{title} Mask')
        plt.imshow(aug_mask, cmap='gray')
        plt.axis('off')

        # Save the images
        aug_img.save(os.path.join(save_folder, f'{title}_Image_{idx}.png'))
        aug_mask.save(os.path.join(save_folder, f'{title}_Mask_{idx}.png'))

    plt.tight_layout()
    plt.show()

# Visualize and save the augmented samples
visualize_and_save_augmented_samples(dataset, 0, save_path)
```
